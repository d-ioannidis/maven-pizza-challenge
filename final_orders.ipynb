```python
import numpy as np
import pandas as pd

from ipynb.fs.full.busiest_times_and_days import *
```

```python
orders = pd.read_csv("D:/github/Python/Python/Maven Pizza Challenge/Resources/pizza_sales/orders.csv", encoding='utf-8')                       
order_details = pd.read_csv("D:/github/Python/Python/Maven Pizza Challenge/Resources/pizza_sales/order_details.csv", encoding='utf-8')
pizza_types = pd.read_csv("D:/github/Python/Python/Maven Pizza Challenge/Resources/pizza_sales/pizza_types.csv", encoding='ISO-8859-1')
pizzas = pd.read_csv("D:/github/Python/Python/Maven Pizza Challenge/Resources/pizza_sales/pizzas.csv", encoding='utf-8')
```

```python
pizza_orders = pd.merge(order_details, pizzas, how='inner', on='pizza_id')
pizza_orders = pd.merge(pizza_orders, pizza_types, how='inner', on='pizza_type_id')
final_orders = pd.merge(pizza_orders, orders, how='inner', on='order_id').drop(
    columns={'ingredients', 'order_details_id', 'pizza_id'}
).sort_values(by='order_id')

final_orders['datetime'] = (final_orders['date'] + ' ' + final_orders['time']).astype('datetime64')
final_orders = final_orders.drop(
        columns={'date', 'time'}
).set_index('datetime')

final_orders['day'] = final_orders.index.day_name()
final_orders['month'] = final_orders.index.month_name()

get_orders_hour_morning(final_orders.index.hour)
get_orders_hour_afternoon(final_orders.index.hour)
get_orders_hour_evening(final_orders.index.hour)
get_orders_hour_night(final_orders.index.hour)

final_orders['time_of_day'] = get_day_category(final_orders.index.hour)
final_orders = final_orders.reset_index()

final_orders["revenue"] = final_orders["price"] * final_orders["quantity"]

final_orders = final_orders.rename(columns={'datetime': 'Datetime', 
'order_id': 'Orders Count', 'quantity': 'Orders Quantity', 
'pizza_type_id': 'Pizza Type', 'size':'Size', 'price': 'Price', 'name': 'Name', 
'category': 'Category', 'day': 'Day', 'month': 'Month', 
'time_of_day': 'Time of Day', 'revenue': 'Revenue'})

final_orders
```

```python
final_orders.to_csv("D:/github/Python/Python/Maven Pizza Challenge/final_orders.csv", encoding='utf-8')
```
